```python
def get_percent(df):
    for movie in df:
        profit_count = df[df['production_budget']*2 <= df['worldwide_gross']] 
        no_profit_count = df[df['production_budget']*2 > df['worldwide_gross']] 
        number_of_movies = len(profit_count) + len(no_profit_count)
        percent_of_pro = len(profit_count)/number_of_movies
        return percent_of_pro
```

```python
print(get_percent(dramadf))
print(get_percent(comedydf))
print(get_percent(thrillerdf))
print(get_percent(actiondf))
print(get_percent(adventuredf))
print(get_percent(horrordf))
print(get_percent(crimedf))
print(get_percent(science_fictiondf))
print(get_percent(romancedf))
print(get_percent(familydf))
```

```python
profitability = [('Drama', 0.433),('Comedy', 0.54),('Thriller', 0.434),('Action', 0.528),('Adventure', 0.583),
                 ('Horror', 0.482),('Crime', 0.397),('Science Fiction', 0.565),('Romance', 0.513),('Family', 0.644)]

profitabilitydf = pd.DataFrame(profitability, columns = ['Genre', 'Percent of Films that were Profitable'])

profitabilitydf['Percent of Films that were Profitable'] = profitabilitydf['Percent of Films that were Profitable']*100

profitabilitydf = profitabilitydf.sort_values(by='Percent of Films that were Profitable', ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(10,12))
plt.xticks(rotation=-45)
plt.rc('axes', labelsize=26) 
plt.rc('xtick', labelsize=14)   
plt.rc('ytick', labelsize=14)  
plt.rc('figure', titlesize=20)
plt.rc('axes', titlesize=26)    
ax.set_ylabel('Genre')
ax.set_title('Percent of Films That Made a Profit by Genre')
ax.set_xlabel('Percent of Films that were Profitable')              
sns.barplot(data=profitabilitydf,y=profitabilitydf['Genre'], x=profitabilitydf['Percent of Films that were Profitable'], orient='h', palette="mako", saturation=.7);
```
